## Bistability Machine Learning Code 
This code is to use machine learning to predict the 
final state of a system of nonlinear differential equations.
The code uses input data generated by solving the differential 
equations in Mathematica. It then trains a neural 
network to predict the steady-state given an initial input.

### - Import modules -

```python
import numpy as np
import tensorflow as tf
import pandas as pd
import random as rnd
import matplotlib.pyplot as plt
import sys 
```

### - Read and scale data -

#### Read in and join all datasets
Note that best practice is not to concat/append the dataframes directly, but instead to create a list of them and then concat. This grows in linear time, rather than concat/append which grows quadratically.

```python
small_dfs = []
    
for i1 in range(21):
    for i2 in range(1,16):
        s1 = str(i1).zfill(2)
        s2 = str(i2).zfill(2)
        #print(s1, " ", s2)
        input_dataset = pd.read_csv("Labelled_SS/"+s1+"_"+s2+".csv")
        small_dfs.append(input_dataset)
dataset=pd.concat(small_dfs, ignore_index=True)
```

```python
dataset_size = dataset.size
print(dataset_size)

data_set_length = len(dataset)
print("Length of dataset = ", data_set_length) 

print(dataset.head())
print(dataset.tail())
```

#### Scale the coherences so between 0 and 1 for later training

```python
dataset[['Re[rho_eg]']] += 0.5
dataset[['Im[rho_eg]']] += 0.5
```

#### Scale the steady state labels for training
Note that this step wouldn't be used if the steady states are classified directly from the data

```python
dataset.loc[dataset['ss'] == 3, 'ss'] = 0
```

#### Classify the steady states from the steady-state coherences and plot
We round to 3 decimal places, but might need to round further if minor errors

```python
df_list=[]
for i1 in range(-10,11):
    for i2 in range(1,16):
        subdata=dataset.loc[(dataset['Delta']==i1) & (dataset['Omega']==i2)].round(3)
        subdata['ss_class']=0
        uniquelist2=np.array([subdata['SSRe[rho_eg]'].unique(), subdata['SSIm[rho_eg]'].unique(), subdata['SSrho_ee'].unique()])
        uniquelist=np.array([len(uniquelist2[0]),len(uniquelist2[1]),len(uniquelist2[2])])
        if np.all(uniquelist == uniquelist[0]):
            ## 1000 entries for each set of params for the ode
            subdata['ss_count']=len(uniquelist2[0])
            ## Classify the steady-state
            for i3 in range(len(uniquelist2[0])):
                subdata.loc[(subdata['SSRe[rho_eg]']==uniquelist2[0][i3]),'ss_class']=i3
            df_list.append(subdata)
        else:
            print("Length error")
            print(i1, " ", i2, " ", subdata['SSRe[rho_eg]'].unique(), " ",subdata['SSIm[rho_eg]'].unique(), " ",subdata['SSrho_ee'].unique())
```

```python
second_data=pd.concat(df_list, ignore_index=True)
newdataset=pd.concat([dataset,second_data[['ss_class','ss_count']]],axis=1)
```

```python
print(newdataset[['ss_count']].value_counts())
print(newdataset[['ss_class']].value_counts())
```

#### Look at number of steady states for each set of parameters

```python
plot_array=newdataset[['Delta','Omega','ss_count']].groupby(['Delta','Omega']).mean().to_numpy()
new_plot_array = plot_array.reshape(21,15)
```

```python
plt.imshow(new_plot_array)
plt.colorbar()
plt.show()
```

#### Look at data imbalance for each set of parameters

```python
## Because the ss_class values are 0 or 1, we can simply sum them for each Delta and Omega
plot_array=newdataset[['Delta','Omega','ss_class']].groupby(['Delta','Omega']).sum().to_numpy()
new_plot_array = plot_array.reshape(21,15)/1000
```

```python
plt.imshow(new_plot_array)
plt.colorbar()
plt.show()
```

### - Analyse the data -
Note that little needs to be done with the data other than scaling carried out above. 
The data is already clean as it has been generated by evaluating the differential equations, so there are no missing values etc. to complete.

#### Rescale Delta and Omega to between 0 and 1 for the training

```python
newdataset[['Omega']] = (newdataset[['Omega']]-1)/14
newdataset[['Delta']] = (newdataset[['Delta']]+10)/20
```

```python
newdataset[['Omega','Delta']]
```

#### Shuffle the dataset, otherwise the validation and test set will be for certain Omega and Delta only

```python
snewdataset=newdataset.sample(frac=1)
print(snewdataset.head())
```

#### Split data into train (60%), validation (20%) and test (20% - drop labels) 

```python
cutone = int(0.6*data_set_length)
cuttwo = int(0.8*data_set_length)

useddataset=snewdataset[['Delta','Omega','Re[rho_eg]','Im[rho_eg]','rho_ee','ss_class']]

dataset_train = useddataset.iloc[:cutone]
dataset_validation = useddataset.iloc[cutone:cuttwo]
dataset_test = useddataset.iloc[cuttwo:]
```

```python
input_length = len(useddataset.columns.values)-1
print(input_length)
```

#### Split the training and validation sets into data and labels for training

```python
X_train = dataset_train.drop('ss_class', axis = 1)
Y_train = dataset_train['ss_class']

X_valid = dataset_validation.drop('ss_class', axis = 1)
Y_valid = dataset_validation['ss_class']

X_test = dataset_test.drop('ss_class', axis = 1)
Y_test = dataset_test['ss_class']
```

### - Compile the model -
Model consists of one input layer with 5 nodes: 2 for the parameters of the differential 
equations and the 3 for the input parameters, then one middle dense layer with 20 nodes and one output layer with 2 nodes for the two steady states.

#### Set number of nodes for middle layer

```python
middlelayernodes = 20 
```

#### Build the model architecture

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(input_length,1)), 
    tf.keras.layers.Dense(middlelayernodes, activation=tf.nn.relu),
    tf.keras.layers.Dense(1, activation='sigmoid')])
```

### - Determine the optimal learning rate -
Train the model for a single epoch with different learning rates to determine a good learning rate for the optimizer

#### Number of learning rates to try

```python
num_learningrates = 100
```

#### Set a lambda function to change the learning rate for each epoch

```python
lr_schedule = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-6 * 10**(epoch / 10))
```

#### Compile the model

```python
model.compile(optimizer = tf.keras.optimizers.Adam(lr=1e-6),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics = ['accuracy'],)
model.summary()
```

#### Train the model

```python
history = model.fit(X_train, Y_train, epochs = num_learningrates, callbacks=[lr_schedule], shuffle=True, batch_size=15)
```

#### Plot the loss vs the learning rate to see optimal value 

```python
plt.semilogx(history.history["lr"], history.history["loss"])
plt.axis([1e-6, 1e3, 0, 1])
plt.xlabel("Learning rate")
plt.ylabel("Error")
plt.show()
```

### - Train model with ideal learning rate -
From the analysis above, a learning rate around 0.01 is good 

#### Clear the previous session 

```python
tf.keras.backend.clear_session()
```

#### Rebuild the architecture 

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(input_length,1)), 
    tf.keras.layers.Dense(middlelayernodes, activation=tf.nn.relu),
    tf.keras.layers.Dense(1, activation='sigmoid') 
    ])
```

#### Compile the model with the optimal learning rate

```python
model.compile(optimizer = tf.keras.optimizers.Adam(lr=1e-2),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics = ['accuracy'],)
model.summary()
```

#### Number of epochs for training

```python
num_epochs = 100
```

#### Fit the model

```python
history = model.fit(X_train, 
                    Y_train, 
                    epochs = num_epochs, 
                    validation_data = (X_valid, Y_valid),
                    shuffle=True, batch_size=15)
```

#### Plot the loss and validation loss in a graph to check for overfitting or underfitting

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(num_epochs)

plt.plot(epochs_range, loss, label = 'Training Loss')
plt.plot(epochs_range, val_loss, label = 'Validation Loss')
plt.legend(loc = 'upper right')
plt.title('Training and Validation Loss')
plt.xlabel("Epoch number")
plt.show()
```

#### Plot the accuracy

```python
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
epochs_range = range(num_epochs)

plt.plot(epochs_range, accuracy, label = 'Training accuracy')
plt.plot(epochs_range, val_accuracy, label = 'Validation accuracy')
plt.legend(loc = 'upper right')
plt.title('Training and Validation accuracy')
plt.xlabel("Epoch number")
plt.show()
```

### - Evaluate the model accuracy -

#### Evaluate just using the error on the test set 

```python
evaluation_train = model.evaluate(X_train, Y_train,)
evaluation_validation = model.evaluate(X_valid, Y_valid,)
evaluation_test = model.evaluate(X_test, Y_test,)
```
